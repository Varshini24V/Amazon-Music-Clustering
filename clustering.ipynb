```python
num_features = [
    'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
    'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms'
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[num_features])
X_scaled
```

```python
X_scaled_df = pd.DataFrame(X_scaled)
X_scaled_df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2,7), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=7)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"K = {k} | Silhouette Score = {score:.4f}")
```

```python
model = KMeans(n_clusters=3)
model.fit(X_scaled)

group = model.predict(X_scaled)
df['cluster'] = group

df
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

score = silhouette_score(X_scaled, df['cluster'])
print("Silhouette Score:", score)
db_score = davies_bouldin_score(X_scaled, df['cluster'])
print("Davies-Bouldin Index:", db_score)
print("Inertia:", kmeans.inertia_)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df['pca1'] = X_pca[:,0]
df['pca2'] = X_pca[:,1]
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x='pca1', y='pca2', hue='cluster', data=df, palette='Set2')
plt.title("Music Clustering using KMeans + PCA")
plt.show()
```

```python
df['cluster'].value_counts()
```

```python
cluster_profile = df.groupby('cluster')[num_features].mean().round(2)
cluster_profile
```

```python
def label_cluster_from_profile(row):
    if row['danceability'] > 0.6 and row['energy'] > 0.6:
        return 'Party Tracks'
    elif row['energy'] < 0.4 and row['acousticness'] > 0.6:
        return 'Chill Acoustic'
    elif row['instrumentalness'] > 0.5:
        return 'Focus / Instrumental'
    elif row['speechiness'] > 0.6:
        return 'Rap / Spoken'
    else:
        return 'Balanced / Mainstream'
```

```python
cluster_profile['cluster_label'] = cluster_profile.apply(label_cluster_from_profile, axis=1)
cluster_profile
```

```python
cluster_label_map = cluster_profile['cluster_label'].to_dict()

df['cluster_label'] = df['cluster'].map(cluster_label_map)
df
```
